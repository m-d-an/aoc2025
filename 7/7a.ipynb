```python
file = open('input.txt', 'r')
```

```python
splitters = []
beam = []
depth = 0
```

```python
for i, line in enumerate(file.readlines()):
    for j, c in enumerate(line.strip()):
        if c == '^':
            splitters.append((i,j))
            depth = i
        if c == 'S':
            beam.append((i,j))
```

```python
def follow_beam(coord):
    #print("follow_beam: "+str(coord))
    splits = 0
    for i in range(coord[0]+1, depth+1):
        if (i,coord[1]) in beam:
            return 0
        else:
            beam.append((i,coord[1]))
        if (i,coord[1]) in splitters:
            splits += follow_beam((i,coord[1]-1))
            splits += follow_beam((i,coord[1]+1))
            return splits + 1
    return splits
```

```python
print(follow_beam(beam[0]))
#example: 21
```
