```python
file = open('input.txt', 'r')
```

```python
splitters = []
beam = []
depth = 0
```

```python
for i, line in enumerate(file.readlines()):
    for j, c in enumerate(line.strip()):
        if c == '^':
            splitters.append((i,j))
            depth = i
        if c == 'S':
            beam.append((i,j))
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def follow_beam(coord):
    #print("follow_beam: "+str(coord))
    timelines = 0
    for i in range(coord[0]+1, depth+1):
        if (i,coord[1]) in splitters:
            timelines += follow_beam((i,coord[1]-1))
            timelines += follow_beam((i,coord[1]+1))
            return timelines + 1
    return timelines
```

```python
print(follow_beam(beam[0])+1)
#example: 21
```
