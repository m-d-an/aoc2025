```python
file = open('input.txt', 'r')
```

```python
#gift_sizes = [7,7,7,7,7,7] # example
gift_sizes = [7,7,7,7,6,5] # input
```

```python
count_feasable = 0
for line in file.readlines():
    if 'x' not in line:
        continue

    #print(line.strip())
    l = line.strip().split(':')

    lw = l[0].split('x')
    length = int(lw[0])
    width = int(lw[1])
    size = length * width
    #print(size)

    gifts = []
    for count in l[1].split():
        gifts.append(int(count))
    #print(gifts)

    min_region_size = 0
    for i, gift_count in enumerate(gifts):
        min_region_size += gift_sizes[i] * gift_count
    #print(min_region_size, size)
    
    if min_region_size + length + width < size:
        count_feasable += 1
print(count_feasable)
```
