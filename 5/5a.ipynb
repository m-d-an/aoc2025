```python
file = open('input.txt', 'r')
```

```python
fresh_ingredients = []
```

```python
sum = 0
for i, line in enumerate(file.readlines()):
    if '-' in line:
        ranges = line.strip().split('-')
        fresh_ingredients.append((int(ranges[0]),int(ranges[1])))
    elif line.strip() != '':
        ingredient = int(line.strip())
        for fir in fresh_ingredients:
            if ingredient >= fir[0] and ingredient <= fir[1]:
                sum += 1
                break
print(sum)
```
