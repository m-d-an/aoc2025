```python
file = open('input.txt', 'r')
```

```python
fresh_ingredients = []
merged_ingredients = []
```

```python
sum = 0
for i, line in enumerate(file.readlines()):
    if '-' in line:
        ranges = line.strip().split('-')
        fresh_ingredients.append((int(ranges[0]),int(ranges[1])))
#print(fresh_ingredients)
```

```python
def insert_into_ingredient_list(ingredients):
    for fir in merged_ingredients:
        lower_overlap = False
        higher_overlap = False
        if ingredients[0] >= fir[0] and ingredients[1] <= fir[1]:
            return False
        if ingredients[0] < fir[0] and ingredients[1] >= fir[0]:
            lower_overlap = True
        if ingredients[0] <= fir[1] and ingredients[1] > fir[1]:
            higher_overlap = True
        #print(ingredients, fir, lower_overlap, higher_overlap)
        if not lower_overlap and not higher_overlap:
            continue
        else:
            merged_ingredients.remove(fir)

        if lower_overlap and not higher_overlap:
            insert_into_ingredient_list((ingredients[0],fir[1]))
            return True
        elif not lower_overlap and higher_overlap:
            insert_into_ingredient_list((fir[0],ingredients[1]))
            return True
        else:
            continue
    merged_ingredients.append(ingredients)
    return True
```

```python

for ingredients in fresh_ingredients:
    insert_into_ingredient_list(ingredients)
```

```python
sum = 0
for ing_ranges in merged_ingredients:
    sum += ing_ranges[1] - ing_ranges[0] + 1
print(sum)
```
