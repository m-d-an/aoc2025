```python
import math
```

```python
file = open('input.txt', 'r')
```

```python
problems = []
```

```python
result = 0
for i, line in enumerate(file.readlines()):
    cols = line.split()
    if cols[0].isdigit():
        for j, c in enumerate(cols):
            if i == 0:
                problems.append([int(c)])
            else:
                problems[j].append(int(c))
    else:
        for j, c in enumerate(cols):
            if c == '*':
                p_result = math.prod(problems[j])
            else:
                p_result = sum(problems[j])
            #print(i, problems[j], p_result)
            result += p_result  
print(result)
# example: 33210 + 490 + 4243455 + 401 = 4277556
```
