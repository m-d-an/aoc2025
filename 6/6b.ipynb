```python
import math
```

```python
file = open('input.txt', 'r')
```

```python
values = [] # lines with numerical values as whole strings
tasks = []
sizes = []
```

```python
for i, line in enumerate(file.readlines()):
    if line[0] != '*' and line[0] != '+': 
        values.append(line[:-1])
    else:
        size = 0
        for c in line[:-1]:
            if c == '*' or c == '+':
                tasks.append(c)
                if size != 0:
                    sizes.append(size)
                    size = 0
            else:
                size += 1
        sizes.append(size+1)
#print(values, tasks, sizes)
```

```python
result = 0
idx_top_left = 0
for i,t in enumerate(tasks):
    numbers = []
    for k in range(sizes[i]):
        number = ""
        for j in range(len(values)):
            number += values[j][idx_top_left+k]
        numbers.append(int(number))
    idx_top_left += sizes[i]+1
    if t == '*':
        result += math.prod(numbers)
    else:
        result += sum(numbers)
print(result)
# example: 1058 + 3253600 + 625 + 8544 = 3263827
```
