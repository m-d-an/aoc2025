```python
file = open('input.txt', 'r')
```

```python
mapping = {}
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def count_paths(node):
    if node == "out":
        return 1
    num_paths = 0
    for connection in mapping[node]:
        num_paths += count_paths(connection)
    return num_paths
```

```python
for i, line in enumerate(file.readlines()):
    spl1 = line.strip().split(':')
    spl2 = spl1[1].strip().split(' ')
    mapping[spl1[0]] = spl2
#print(mapping)
paths = count_paths("you")
print(paths)
```
