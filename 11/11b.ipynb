```python
file = open('input.txt', 'r')
```

```python
mapping = {}
```

```python
from functools import lru_cache

@lru_cache(maxsize=None)
def count_paths_to_out(node,dac,fft):
    if node == "out":
        if dac and fft:
            return 1
        else:
            return 0
    if node == "dac":
        dac = True
    if node == "fft":
        fft = True
    num_paths = 0
    for connection in mapping[node]:
        num_paths += count_paths_to_out(connection,dac,fft)
    return num_paths
```

```python
for i, line in enumerate(file.readlines()):
    spl1 = line.strip().split(':')
    spl2 = spl1[1].strip().split(' ')
    mapping[spl1[0]] = spl2
#print(mapping)
```

```python
paths_out = count_paths_to_out("svr",False,False)
print(paths_out)
```
