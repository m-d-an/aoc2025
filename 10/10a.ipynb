```python
file = open('input.txt', 'r')
```

```python
from itertools import combinations

fewest_presses = 0
for line in file.readlines():
    s = line.strip().split(' ')
    lights_target = []
    for c in s[0]:
        if c == '.':
            lights_target.append(False)
        elif c == '#':
            lights_target.append(True)
    #print(lights_target)

    buttons = []
    for b in s[1:-1]:
        idxs = b[1:-1].split(',')
        for i in range(len(idxs)):
            idxs[i] = int(idxs[i])
        buttons.append(tuple(idxs))
    #print(buttons)

    combs = []
    for r in range(len(buttons) + 1):
        combs.extend(combinations(buttons, r))
    #print(combs)

    for c in combs:
        lights = [False] * len(lights_target)
        for b in c:
            for li in b:
                lights[li] = not lights[li]
        if lights == lights_target:
            #print(c, lights)
            fewest_presses += len(c)
            break
    #print()
print(fewest_presses)
```
