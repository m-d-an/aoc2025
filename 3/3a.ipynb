```python
file = open('input.txt', 'r')
```

```python
sum = 0
for line in file.readlines():
    joltage_ratings = []
    for c in line.strip():
        joltage_ratings.append(int(c))
    #print(joltage_ratings)
    idx_1 = joltage_ratings.index(max(joltage_ratings))
    val_1 = joltage_ratings[idx_1]
    if len(joltage_ratings[idx_1+1:]) > 0:
        val_2 = max(joltage_ratings[idx_1+1:])
        joltage = val_1*10+val_2
    else:
        joltage_ratings.pop(idx_1)
        val_2 = max(joltage_ratings)
        joltage = val_2*10+val_1
    #print(joltage)
    sum += joltage
print(sum)
#example: 98 + 89 + 78 + 92 = 357
```
