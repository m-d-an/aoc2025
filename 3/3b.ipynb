```python
file = open('input.txt', 'r')
```

```python
joltage_ratings = []
```

```python
def get_max(index, reserve):
    if reserve == 0:
        idx = joltage_ratings[index:].index(max(joltage_ratings[index:]))
    else:
        idx = joltage_ratings[index:-(reserve)].index(max(joltage_ratings[index:-(reserve)]))
    val = joltage_ratings[index + idx]
    return index + idx + 1, val
```

```python
sum = 0
for line in file.readlines():
    joltage_ratings = []
    joltage = 0
    idx = 0
    for c in line.strip():
        joltage_ratings.append(int(c))
    #print(joltage_ratings)
    for i in range(12):
        idx, val = get_max(idx, 11-i)
        joltage = joltage * 10 + val
    #print(joltage)
    sum += joltage
print(sum)
#example: 987654321111 + 811111111119 + 434234234278 + 888911112111 = 3121910778619
```
