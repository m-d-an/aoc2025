```python
file = open('input.txt', 'r')
```

```python
junction_points = []
distances = []
```

```python
for i, line in enumerate(file.readlines()):
    points = line.strip().split(',')
    junction_points.append((int(points[0]),int(points[1]),int(points[2])))
```

```python
import math

for i in range(len(junction_points)):
    for j in range(i+1, len(junction_points)):
        d = math.dist(junction_points[i], junction_points[j])
        distances.append((d, i, j))
distances.sort()
```

```python
circuits = []
for jp in junction_points:
    circuits.append([jp])
```

```python
i = 0
while True:
    shortest = distances[i]
    i += 1
    box1 = junction_points[shortest[1]]
    box2 = junction_points[shortest[2]]
    included = []
    for j, circuit in enumerate(circuits):
        if box1 in circuit:
            included.append(j)
        if box2 in circuit:
            included.append(j)
    if included[0] != included[1]:
        circuits.append(list(set( circuits[included[0]] + circuits[included[1]])))
        if included[0] > included[1]:
            circuits.pop(included[0])
            circuits.pop(included[1])
        else:
            circuits.pop(included[1])
            circuits.pop(included[0])
    if len(circuits) == 1:
        print(box1[0]*box2[0])
        break
```
