```python
file = open('input.txt', 'r')
```

```python
tiles = []
```

```python
for line in file.readlines():
    s = line.strip().split(',')
    tiles.append((int(s[0]),int(s[1])))
```

```python
max_area = 0
for t1 in tiles:
    for t2 in tiles:
        area = abs((t1[0]-t2[0]+1) * (t1[1]-t2[1]+1))
        if area > max_area:
            max_area = area
print(max_area)
```
