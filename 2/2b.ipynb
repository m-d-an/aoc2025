```python
file = open('input.txt', 'r')
```

```python
ranges = file.readlines()[0].split(',')
sum = 0
for r in ranges:
    start = int(r.split('-')[0])
    end = int(r.split('-')[1])
    for i in range(start,end+1):
        l = len(str(i))
        for j in range(1,int(l/2)+1):
            if l % j != 0:
                continue
            else:
                valid = False
                for k in range(0,int(l/j)-1):
                    a = str(i)[k*j:(k+1)*j]
                    b = str(i)[(k+1)*j:(k+2)*j]
                    if a != b:
                        valid = True    
                        break
                if not valid:
                    #print(i)
                    sum += i
                    break
print(sum)
# example: 4174379265
```
