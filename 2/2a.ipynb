```python
file = open('input.txt', 'r')
```

```python
number = 50
zeroes = 0
for line in file.readlines():
    if line[0] == 'L':
        number -= int(line.strip()[1:])
    else:
        number += int(line.strip()[1:])
    number = number % 100
    if number == 0:
        zeroes += 1
    #print(number)
print(zeroes)
```
